## 【問題1】乱数の作成
参考URL  
http://starpentagon.net/analytics/numpy_norm_rand/

```python
import numpy as np

mu1 = [-3, 0] #平均μ
sigma1 = [[1.0, 0.8], [0.8, 1.0]]#共分散行列Σ
np.random.seed(0) #シード値0

values1 = np.random.multivariate_normal(mu1, sigma1, 500)
print(values1)
```

## 【問題2】散布図による可視化
参考URL  
https://note.nkmk.me/python-numpy-select-element-row-column-array/

```python
import matplotlib.pyplot as plt
x = values1[:, 0]
y = values1[:, 1]

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

plt.scatter(x, y)
plt.show()
```

## 【問題3】ヒストグラムによる可視化

```python
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")

plt.hist(y, bins = 50)
plt.xlim(-5.9, 2.9)
plt.show()
```

## 【問題4】データの追加
参考URL  
https://paper.hatenadiary.jp/entry/2017/05/02/152223

```python
mu4 = [0, -3] #平均μ
sigma4 = [[1.0, 0.8], [0.8, 1.0]]#共分散行列Σ
np.random.seed(0) #シード値0

values4 = np.random.multivariate_normal(mu4, sigma4, 500)
print(values4)
```

```python
x = values1[:, 0]
y = values1[:, 1]

x4 = values4[:, 0]
y4 = values4[:, 1]

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

plt.scatter(x, y, label = "0")
plt.scatter(x4, y4, label = "1")
plt.legend(loc='upper right')
plt.show()
```

## 【問題5】データの結合

```python
toi5 = np.concatenate([values1, values4])
print(toi5.shape)
```

## 【問題6】ラベル付け
参考URL  
https://note.nkmk.me/python-numpy-append/

```python
zero = np.zeros((500, 1))
one = np.ones((500, 1))
zero_one = np.concatenate([zero, one])
toi6 = np.append(toi5, zero_one, axis = 1)
print(toi6.shape)
print(toi6)
```
